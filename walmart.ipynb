```python
def walmart_product_id(keyword):
    
    """
    Function:
    ---------
        (1) walmart_product_id pulls all the product id for products which have reviews

    Args:
    -----
        (1) keyword (str): search term defined by the user

    Returns:
    --------
        (1) product_id_list (list): list of product id
    """
    
    # find the maximum page of search page
    url = 'https://www.walmart.com/search/?query=' + keyword.replace(" ", "%20")
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, 'html.parser')
    
    max_page = int(page_soup.find("div",{"class":"paginator"}).ul.li.a.get('aria-label').split()[-2])
    print('Found {} search pages'.format(max_page))
    
    # get all the search pages url
    p = 1
    search_url = []
    while p <= max_page:
        url = 'https://www.walmart.com/search/?page=' + str(p) + '&ps=40&query='  + keyword.replace(" ", "+") + '#searchProductResult'
        search_url.append(url)
        p += 1
    
    # get all the product id
    product_id_list = []
    for u in range(len(search_url)):
        uClient = uReq(search_url[u])
        page_html = uClient.read()
        uClient.close()
        page_soup = soup(page_html, 'html.parser')
        
        product_id_raw = page_soup.findAll("a",{"class":"product-title-link line-clamp line-clamp-2"})
        product_id_review_count_raw = page_soup.findAll("span",{"class":"seo-review-count"})
        
        length = 0
        for j in range(len(product_id_raw)):
            product_id_temp = product_id_raw[j].get('href').split('/')[3].split('?')[0]
            product_id_review_count_temp = int(product_id_review_count_raw[j].string)

            # remove repeated product id and product without reviews
            if product_id_temp not in product_id_list and product_id_review_count_temp != 0: 
                product_id_list.append(product_id_temp)
                length += 1
        print('Found {} products with reviews in search page {} of {}'.format(length, u+1, len(search_url)))
    
    print('Found {} distinct products with reviews in total'.format(len(product_id_list)))
    print(product_id_list) ##############
    
    return product_id_list
```

```python
def walmart_review_scraper(product_id):
    """
    Function:
    ---------
        (1) walmart_review_scraper pulls the relevant information for a specific product   

    Args:
    -----
        (1) product_id (str): unique identifier for the product 

    Returns:
    --------
        (1) reviews_df (pandas DataFrame): pandas DataFrame with the following columns:
            (a) Source
            (b) Brand
            (c) Product ID
            (d) Name
            (e) Date
            (f) Rating
            (g) Usefulness
            (h) Review Title
            (i) Review
    """
    
    url = 'https://www.walmart.com/reviews/product/' + str(product_id)
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, 'html.parser')
    
    # get product name
    product_name = page_soup.find('div', {'class':"product-summary-wrapper"}).a.h1.div.div.text

    # get dates
    date_raw = page_soup.findAll('span', {'itemprop':"datePublished"})
    date = []
    for l in range(len(date_raw)):
        date.append(date_raw[l].text)

    # get star ratings
    rating_raw = page_soup.findAll("div",{"class":"stars stars-small"})
    rating = []
    for j in range(len(rating_raw)):
        rating.append(rating_raw[j].span["alt"][16]) 
    
    # get usefulness
    usefulness_raw = page_soup.findAll('span', {'class':"yes-no-count"})[::2]
    usefulness = []
    for u in range(len(usefulness_raw)):
        usefulness.append(usefulness_raw[u].text) 
        
    # get review titles
    review_title_raw = page_soup.findAll("div",{"class":"Grid review-heading"})
    review_title = []
    for i in range(len(review_title_raw)):
        if review_title_raw[i].find("div",{"class":"review-title"}) == None:
            review_title.append(" ")
        else:
            review_title.append(review_title_raw[i].find("div",{"class":"review-title"}).text)

    # get review details
    review_details_raw = page_soup.findAll("div",{"class":"review-body-text"})
    review_details = []
    for k in range(len(review_details_raw)):
        review_details.append(review_details_raw[k].text.replace('\n\n',' ').replace('\n',' '))
                  
    # get brand
    brand_url = "https://www.walmart.com/ip/" + str(product_id)
    uClient = uReq(brand_url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, 'html.parser')
    brand = page_soup.find("span",{"itemprop":"brand"}).text
    
    # Consolidate information into a pandas DataFrame
    reviews_df = pd.DataFrame()
    for m in range(len(date)):
        review_dict = { 'Source': "Walmart",
                       'Brand': brand,
                       'Product ID': product_id,
                       'Name': product_name,
                       'Date': date[m],
                       'Rating': rating[m],
                       'Usefulness': usefulness[m],
                       'Review Title': review_title[m],
                       'Review': review_details[m]}
        reviews_df = reviews_df.append(review_dict, ignore_index=True)
        
    return reviews_df
```

```python
def walmart_scrape_to_df(keyword):
    """
    Function:
    ---------
        (1) walmart_scrape_to_df iterates through all the product id to pulls the relevant information

    Args:
    -----
        (1) keyword (str): search term defined by the user

    Returns:
    --------
        output_df (pandas DataFrame): pandas DataFrame with the following columns:
            (a) Source
            (b) Brand
            (c) Product ID
            (d) Name
            (e) Date
            (f) Rating
            (g) Usefulness
            (h) Review Title
            (i) Review
    """
    
    # gather all the product id
    product_id_list = walmart_product_id(keyword)
    
    output_df = pd.DataFrame()
    for p in range(len(product_id_list)):
        print("Scraping reviews for {}, {} of {} products.".format(product_id_list[p], p+1, str(len(product_id_list))))
        reviews_df = walmart_review_scraper(product_id_list[p])
        output_df = output_df.append(reviews_df, ignore_index = True)
    
    return output_df
```
