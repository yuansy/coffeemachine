```python
from google.cloud import translate_v3beta1 as translate
client = translate.TranslationServiceClient()
```

```python
def google_translate(text, source='pt', target='en'):

    project_id = 'portuguese-1561102534379'
    text = text
    location = 'global'

    parent = client.location_path(project_id, location)

    response = client.translate_text(
        parent=parent,
        contents=[text],
        mime_type='text/plain',  # mime types: text/plain, text/html
        source_language_code=source,
        target_language_code=target)

    # for translation in response.translations:
    #     print('Translated Text: {}'.format(translation))

    return str(response.translations[0])[18:-2]
```

```python
import pandas as pd
```

```python
df = pd.read_excel('glassdoor_brf_raw.xls')
```

```python
df_preprocess = pd.DataFrame()
df_preprocess['reviews_raw'] = df.PROs.tolist() + df.CONs.tolist()
df_preprocess['language'] = df.language.tolist()*2
```

```python
df_preprocess
```

```python
translate = []
for r,l in zip(df_preprocess.reviews_raw, df_preprocess.language):
    if l == 'e':
        translate.append(r)
    elif l == 'p':
        translate.append(google_translate(r))

df_preprocess['reviews'] = translate
```
